```python
# Importar las bibliotecas necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Suponiendo que 'df' es tu dataframe original
metrics_df = df[['daily_rank', 'snapshot_date', 'popularity', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature']]

# Convertir 'snapshot_date' a datetime si no lo está ya
metrics_df['snapshot_date'] = pd.to_datetime(metrics_df['snapshot_date'])

# Calcular la matriz de correlación (excluyendo la columna de fecha)
correlation_matrix = metrics_df.drop(columns=['snapshot_date']).corr()

# Crear el mapa de calor
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title('Mapa de Correlación')
plt.show()
```
